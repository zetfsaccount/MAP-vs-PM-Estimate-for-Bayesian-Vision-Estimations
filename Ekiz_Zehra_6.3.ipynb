1) (Based on Problem 6.3) In Section 6.1, we noted that the probability of rain the next day was given by the posterior mean estimate, and was different from the most probable value of the rainfall rate, r, on the island, which is the MAP estimate. Suppose you’ve been on the island for 4 days, and it has rained only once.

a) Show that the MAP estimate for r is 0.25 and, according to Laplace’s rule of succession (Eq. (6.14)), the posterior mean (PM) estimate is 0.33.

In this question number of rained day is 1 while total time passed is 4.
For PM, we use the formula of rainy day plus one over time plus two ((1+1)/(4+2))=0.33 while +we use the formula of rainy day over time for MAP (1/4)=0.25. It is because while posterior estimate is regarding prediction, MAP is about the probability of world state.

b) Verify Eq. (6.14) numerically, by discretizing r into several hundred or thousand values equally spaced between 0 and 1. Given 1 rainy day out of 4, calculate the likelihood for each r value, and enter these into Bayes’ formula with a uniform prior. Calculate the posterior mean as ∑ r p(r|xobs). Show that it is equal to the value obtained using the formula.

b) Verify Eq. (6.14) numerically, by discretizing r into several hundred or thousand values equally spaced between 0 and 1. Given 1 rainy day out of 4, calculate the likelihood for each r value, and enter these into Bayes’ formula with a uniform prior. Calculate the posterior mean as ∑ r p(r|xobs). Show that it is equal to the value obtained using the formula.

```python

import numpy as np
import matplotlib.pyplot as plt
import random


r_array = np.arange(0, 1.001, 0.001)

l_rainy = r_array
l_dry_1 = 1-r_array
l_dry_2 = 1-r_array
l_dry_3 = 1-r_array
posterior = l_rainy*l_dry_1*l_dry_2*l_dry_3
posterior = posterior/sum(posterior)


plt.figure()
plt.plot(r_array,l_rainy,color='blue')
plt.plot(r_array,l_dry_1,color='red')
plt.plot(r_array,l_dry_2,color='green')
plt.plot(r_array,l_dry_3,color='yellow')
plt.title("Likelihood")
plt.xlabel("Prob. of Rain r")
plt.ylabel("Likelihood")

plt.figure()
plt.plot(r_array,posterior)
plt.title("Posterior")
plt.xlabel("Prob. of Rain r")
plt.ylabel("Posterior")

pm_estimate = sum(r_array*posterior)
print("PM estimate: {:.5f}".format(pm_estimate))
```

c) Imagine that you repeat this 4-day experiment 1000 times (for example you live this 4-day period in your life again and again as in a sci-fi movie). In each 4-day experiment, the probability of rain is r = 0.25. Simulate the experiments, where rain every day is the outcome of a Bernoulli process with probability r. For each experiment compute the MAP and PM estimates of r using the formulas. Next, compute the means of MAP and PM estimates across the 1000 experiments. Which one is closer to the true r?

MAP is closer to r

```python
r = 0.25
pm_estimate = []
map_estimate = []
for i in range(1000):
    sampleList = [1, 0]
    randomList = random.choices(sampleList, weights=(r, 1-r), k=4)
    pm_estimate.append((sum(randomList)+1)/(4+2))
    map_estimate.append(sum(randomList)/4)

mean_pm = sum(pm_estimate)/len(pm_estimate)
mean_map = sum(map_estimate)/len(map_estimate)
print("For r = 0.25")
print("Mean PM: {:.5f}".format(mean_pm))
print("Mean MAP: {:.3f}".format(mean_map))
print()
```

f) Interpret your results in (c)-(e). Why does this pattern make intuitive sense?

It makes intuitive sense since the patterns are since for the MAP we only consider probability as a whole while we consider the contribution of the day number we are predicting for PM. If we consider the discrepancy between values of MAP and PM for each condition of r, this is logical as while r=0.25, like one day is rained but 3 days were dry, the direct calculation of prob. is 1/4 as seen on MAP. However, PM is 0.33 since it also considers the day we are predicting while calculating the probability. While MAP is more near to r, in real world PM makes more intuitive sense since the day we are predicting is also affective to model progress in learning. For 0.5 both MAP and PM equal since the prob. is kind of flat on former experiences.
